## **Data Pre-processing**

```python
data.info()
```

```python
data.describe()
```

```python
data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].mean(),inplace = True)
data['Revenue (Millions)'].tail()
```

```python
data[data['Revenue (Millions)']<1]
```

```python
data.drop((data[data['Revenue (Millions)']<1]).index,inplace =True)
```

## **Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print('Following are number of movies released per year')
print(data.Year.value_counts())
plt.title("Movies released per year")
sns.countplot(x=data['Year'],data =data);
```

Most movies realease in 2016

```python
plt.figure(figsize=(8,6))
sns.scatterplot(y='Revenue (Millions)',x=data['Votes'],data=data,color='green');
plt.title("Revenue generated v/s votes obtained");
```

The plot is clustered near origin, it can be assumed out with the increasing Revenue, the number of votes hardly increase. Also, there are very few movies with very high revenue.

```python
sns.scatterplot(y='Revenue (Millions)',x=data['Votes'],data=data.loc[data['Revenue (Millions)']<400],color='pink');
```

This concludes, that Revenue generated and votes earned by the movies are not linearly realated.

```python
sns.histplot(x='Runtime (Minutes)',data=data,color='red');
plt.title('No.of movies as per their runtime');
```

This is bell shape curve, that means most of the movies runtime is around the median,that is 80 minutes to 120 minutes duration.

```python
data.groupby("Year")['Revenue (Millions)'].mean()
```

Highest revenue was generated in year 2009 which was ~112 Millions

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=data,color='skyblue');
plt.title('Revenue generated per year');
```
